```python
import random
import os
seed = 42
random.seed(seed)
np.random.seed(seed)
os.environ["PYTHONHASHSEED"] = str(seed)
```

```python
df = dict()
df['2015'] = pd.read_csv('/kaggle/input/-')
df['2016'] = pd.read_csv('/kaggle/input/-')
df['2017'] = pd.read_csv('/kaggle/input/-')
df['2018'] = pd.read_csv('/kaggle/input/-')
df['2019'] = pd.read_csv('/kaggle/input/-')
df['2020'] = pd.read_csv('/kaggle/input/-')

tf = dict()
tf['2021'] = pd.read_csv('/kaggle/input/-')
tf['2022'] = pd.read_csv('/kaggle/input/-')

submit = pd.read_csv('/kaggle/input/-')
print(df['2015'].info())
print(df['2016'].info())
print(df['2017'].info())
print(df['2018'].info())
print(df['2019'].info())
print(df['2020'].info())

print(submit.info())
```

```python
# 필요한 데이터만 추출 ['GDP per capita', 'Healthy Life Expectancy', 'Social support', 'Freedom to make life choices', 'Generosity', 'Corruption Perception', 'Residual error']

train_col = ['GDP per capita', 'Social support', 'Healthy Life Expectancy', 'Freedom to make life choices', 'Generosity', 'Corruption Perception', 'Residual error', 'Happiness Score']
test_col = ['GDP per capita', 'Social support', 'Healthy Life Expectancy', 'Freedom to make life choices', 'Generosity', 'Corruption Perception', 'Residual error']

df['2015'] = df['2015'][train_col]
df['2016'] = df['2016'][train_col]
df['2017'] = df['2017'][train_col]
df['2018'] = df['2018'][train_col]
df['2019'] = df['2019'][train_col]
df['2020'] = df['2020'][train_col]

tf['2021'] = tf['2021'][test_col]
tf['2022'] = tf['2022'][test_col]


```

```python
# 필요한 데이터 추출 및 병합과정(concat) -> index reset을 통해 오류가능성 제거
df['2018']['Residual error'] =  df['2018']['Happiness Score'] - (df['2018']['GDP per capita'] + df['2018']['Social support'] + df['2018']['Healthy Life Expectancy'] + df['2018']['Freedom to make life choices'] + df['2018']['Generosity'] + df['2018']['Corruption Perception'])
df['2019']['Residual error'] =  df['2019']['Happiness Score'] - (df['2019']['GDP per capita'] + df['2019']['Social support'] + df['2019']['Healthy Life Expectancy'] + df['2019']['Freedom to make life choices'] + df['2019']['Generosity'] + df['2019']['Corruption Perception'])


train_data = pd.concat([df['2015'],df['2016'],df['2017'],df['2018'],df['2019'],df['2020']],axis=0)
test_data = pd.concat([tf['2021'],tf['2022']],axis=0)

train_data = train_data.reset_index(drop=True)
test_data = test_data.reset_index(drop=True)

test_data
```

```python
train_data.info()
test_data.info()
```

```python
# train 데이터는 float 형으로 제대로 들어가있는걸 확인했지만, 결측치가 존재할 수 있으니 최빈값 처리함 (pandas 함수의 fillna 활용)

for col in train_data.columns:
    if train_data[col].isnull().sum() != 0:
        train_data[col] = train_data[col].fillna(train_data[col].mode()[0])
```

```python
# test data 확인결과 소숫점이 , 으로 표현된것을 확인 이를 소수로 수정함

for col in test_data.columns:
    for i in range(len(test_data[col])):
        if type(test_data[col][i]) != float:
            test_data[col][i] = test_data[col][i].replace(',','.')
            test_data[col][i] = float(test_data[col][i])
    test_data[col] = test_data[col].astype('float64')

print(test_data.info())
```

```python
print(train_data.info())
x_data = train_data.iloc[:,:-1].to_numpy()
y_data = train_data.iloc[:, -1].to_numpy()
test_data = test_data.to_numpy()

from sklearn.preprocessing import StandardScaler
std = StandardScaler()

x_data_std = std.fit_transform(x_data)
test_data_std = std.transform(test_data)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(x_data_std, y_data)
pred = knn.predict(test_data_std)
pred
```

```python
submit['score'] = pred
submit.to_csv('submit', header=True, index=False)
```
