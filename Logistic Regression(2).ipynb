```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session

import random

seed = 42

random.seed(seed)
np.random.seed(seed)
os.environ["PYTHONHASHSEED"] = str(seed)
```

```python
train_X = pd.read_csv('/kaggle/input/2023-ml-w5p2/trainX.csv')
print(train_X.head())
train_Y = pd.read_csv('/kaggle/input/2023-ml-w5p2/trainY.csv')
print(train_Y.head())
test_X = pd.read_csv('/kaggle/input/2023-ml-w5p2/testX.csv')
print(test_X.head())
submit = pd.read_csv('/kaggle/input/2023-ml-w5p2/submit.csv')
print(submit.head())
```

```python
train_Y = train_Y['label']
print(train_Y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_X, train_Y, test_size=0.3, random_state=100, stratify=train_Y)
```

```python
from sklearn.linear_model import LogisticRegression

normal = LogisticRegression()

l2_100 = LogisticRegression(penalty = 'l2', C= 100.0, solver = 'lbfgs', max_iter= 1000000)
l2_10 = LogisticRegression(penalty='l2', C=10.0, solver='saga', max_iter = 1000000)  # L2 with C(=1/λ)=10
l2_1 = LogisticRegression(penalty='l2', C=1.0, solver='saga', max_iter = 1000000)    # L2 with C(=1/λ)=1
l2_0_1 = LogisticRegression(penalty='l2', C=0.1, solver='saga', max_iter = 1000000)  # L2 with C(=1/λ)=0.1

l1_10 = LogisticRegression(penalty='l1', C=10.0, solver='saga', max_iter = 1000000)  # L1 with C(=1/λ)=10
l1_1 = LogisticRegression(penalty='l1', C=1.0, solver='saga', max_iter = 1000000)    # L1 with C(=1/λ)=1
l1_0_1 = LogisticRegression(penalty='l1', C=0.1, solver='saga', max_iter = 1000000)  # L1 with C(=1/λ)=0.1
```

```python
# 정규화 안하는게 더 나음
normal.fit(X_train, y_train)
print('학습데이터 정밀도 with all baseline:', normal.score(X_train, y_train))
print('시험데이터 정밀도 with all baseline:', normal.score(X_test, y_test))

l2_100.fit(X_train, y_train)
print('학습데이터 정밀도 with L2 and λ=0.01:', l2_100.score(X_train, y_train))
print('시험데이터 정밀도 with L2 and λ=0.01:', l2_100.score(X_test, y_test))


l2_10.fit(X_train, y_train)
print('학습데이터 정밀도 with L2 and λ=0.1:', l2_10.score(X_train, y_train))
print('시험데이터 정밀도 with L2 and λ=0.1:', l2_10.score(X_test, y_test))

l2_1.fit(X_train, y_train)
print('학습데이터 정밀도 with L2 and λ=1:', l2_1.score(X_train, y_train))
print('시험데이터 정밀도 with L2 and λ=1:', l2_1.score(X_test, y_test))

l2_0_1.fit(X_train, y_train)
print('학습데이터 정밀도 with L2 and λ=10:', l2_0_1.score(X_train, y_train))
print('시험데이터 정밀도 with L2 and λ=10:', l2_0_1.score(X_test, y_test))

l1_10.fit(X_train, y_train)
print('학습데이터 정밀도 with L1 and λ=0.1:', l1_10.score(X_train, y_train))
print('시험데이터 정밀도 with L1 and λ=0.1:', l1_10.score(X_test, y_test))

l1_1.fit(X_train, y_train)
print('학습데이터 정밀도 with L1 and λ=1:', l1_1.score(X_train, y_train))
print('시험데이터 정밀도 with L1 and λ=1:', l1_1.score(X_test, y_test))

l1_0_1.fit(X_train, y_train)
print('학습데이터 정밀도 with L1 and λ=10:', l1_0_1.score(X_train, y_train))
print('시험데이터 정밀도 with L1 and λ=10:', l1_0_1.score(X_test, y_test))
```

```python
print(l2_100.intercept_)
print(l2_10.intercept_)
print(l2_1.intercept_)
print(l2_0_1.intercept_)

print(l2_100.coef_)
print(l2_10.coef_)
print(l2_1.coef_)
print(l2_0_1.coef_)
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
X_train_std = std.fit_transform(X_train)
X_test_std = std.transform(X_test)
```
