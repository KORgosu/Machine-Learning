```python
train_data = pd.read_csv('/kaggle/input/2023-ml-w7p1/train.csv')
test_data = pd.read_csv('/kaggle/input/2023-ml-w7p1/test.csv')
submit = pd.read_csv('/kaggle/input/2023-ml-w7p1/sample_submit.csv')

print(train_data.head())
print(test_data.head())
print(submit.head())
```

```python
x_data = train_data.drop(['index','enrollee_id', 'target', 'city'], axis=1)
y_data = train_data['target']
test_data = test_data.drop(['index', 'enrollee_id', 'city'], axis=1)

print(x_data.head())
print(y_data.head())
print(test_data.head())

```

```python
from sklearn.preprocessing import LabelEncoder



LE = LabelEncoder()
x_data_LE = LE.fit_transform(x_data['gender'].values)
test_data_LE = LE.transform(test_data['gender'].values)
x_data['gender'] = x_data_LE
test_data['gender'] = test_data_LE

LE1 = LabelEncoder()
x_data_LE = LE1.fit_transform(x_data['relevent_experience'].values)
test_data_LE = LE1.transform(test_data['relevent_experience'].values)
x_data['relevent_experience'] = x_data_LE
test_data['relevent_experience'] = test_data_LE

LE2 = LabelEncoder()
x_data_LE = LE2.fit_transform(x_data['enrolled_university'].values)
test_data_LE = LE2.transform(test_data['enrolled_university'].values)
x_data['enrolled_university'] = x_data_LE
test_data['enrolled_university'] = test_data_LE

LE3 = LabelEncoder()
x_data_LE = LE3.fit_transform(x_data['education_level'].values)
test_data_LE = LE3.transform(test_data['education_level'].values)
x_data['education_level'] = x_data_LE
test_data['education_level'] = test_data_LE

LE4 = LabelEncoder()
x_data_LE = LE4.fit_transform(x_data['major_discipline'].values)
test_data_LE = LE4.transform(test_data['major_discipline'].values)
x_data['major_discipline'] = x_data_LE
test_data['major_discipline'] = test_data_LE

LE5 = LabelEncoder()
x_data_LE = LE5.fit_transform(x_data['company_type'].values)
test_data_LE = LE5.transform(test_data['company_type'].values)
x_data['company_type'] = x_data_LE
test_data['company_type'] = test_data_LE

print(x_data.head())
print(test_data.head())
```

```python
# >20 -> 21로, <1 -> 0으로 nan -> -1로 변경
XX = x_data['experience'].copy()
TT = test_data['experience'].copy()
XX[XX=='>20']=21
XX[XX=='<1']=0
XX[XX.isnull()]=-1

TT[TT=='>20']=21
TT[TT=='<1']=0
TT[TT.isnull()]=-1

x_data['experience'] = XX.astype('int64')
x_data['experience'].unique()
```

```python
test_data['experience'] = TT.astype('int64')
test_data['experience'].unique()
```

```python
x_data['company_size'].unique()
```

```python
XX = x_data['company_size'].copy()
XX[XX=='10000+']=1
XX[XX=='5000-9999']=2
XX[XX=='1000-4999']=3
XX[XX=='500-999']=4
XX[XX=='100-500']=5
XX[XX=='50-99']=6
XX[XX=='10/49']=7
XX[XX=='<10']=8
XX[XX.isnull()]=-1

x_data['company_size'] = XX.astype('int64')
x_data['company_size'].unique()
```

```python
TT = test_data['company_size'].copy()
TT[TT=='10000+']=1
TT[TT=='5000-9999']=2
TT[TT=='1000-4999']=3
TT[TT=='500-999']=4
TT[TT=='100-500']=5
TT[TT=='50-99']=6
TT[TT=='10/49']=7
TT[TT=='<10']=8
TT[TT.isnull()]=-1

test_data['company_size'] = TT.astype('int64')
test_data['company_size'].unique()
```

```python
XX = x_data['last_new_job'].copy()
XX[XX=='>4']=5
XX[XX=='never']=0
XX[XX.isnull()]=-1
x_data['last_new_job'] = XX.astype('int64')
x_data['last_new_job'].unique()

```

```python
TT = test_data['last_new_job'].copy()
TT[TT=='>4']=5
TT[TT=='never']=0
TT[TT.isnull()]=-1
test_data['last_new_job'] = TT.astype('int64')
test_data['last_new_job'].unique()
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda  = QuadraticDiscriminantAnalysis()
qda.fit(x_data, y_data)
model = qda.predict(test_data)

submit['target'] = model.astype('int64')
submit.to_csv('submit.csv', header=True, index=False)
print(submit.head())
```
