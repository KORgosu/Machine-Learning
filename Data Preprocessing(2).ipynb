```python
train_data = pd.read_csv('/kaggle/input/2023-ml-w7p2/train.csv')
test_data = pd.read_csv('/kaggle/input/2023-ml-w7p2/test.csv')
submit = pd.read_csv('/kaggle/input/2023-ml-w7p2/sample_submit.csv')

print(train_data.info())
print(test_data.info())
print(submit.head())
```

```python
test_data = test_data.drop(['index', 'customerID'], axis=1)

x_data = train_data.drop(['index', 'Churn', 'customerID'], axis=1)
y_data = train_data['Churn']

print(x_data.info())
print(test_data.info())

print(x_data['TotalCharges'][280])
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
col = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'TotalCharges']

for columns in col:
    x_data[columns] = LE.fit_transform(x_data[columns].values)

col = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']
for columns in col:
    test_data[columns] = LE.fit_transform(test_data[columns].values)

y_data = LE.fit_transform(y_data)
print(x_data.info())
print(test_data.info())
```

```python
x_data['TotalCharges'] = x_data['TotalCharges'].astype('float64')

print(x_data.info())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.3, random_state = 100, stratify=y_data)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression(max_iter=1000)
reg.fit(X_train, y_train)
pred = reg.predict(X_test)

from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, pred))
```

```python
reg.fit(x_data, y_data)
model = reg.predict(test_data)
model

submit['Churn'] = model
submit.to_csv('submit.csv', header=True,index=False)
```
