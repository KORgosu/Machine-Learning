```python
x_data = pd.read_csv('/kaggle/input/2023-ml-w6p1/train.csv')
print(x_data.head())
y_data = pd.read_csv('/kaggle/input/2023-ml-w6p1/train_label.csv')
print(y_data.head())
test_data = pd.read_csv('/kaggle/input/2023-ml-w6p1/test.csv')
print(test_data.head())
submit = pd.read_csv('/kaggle/input/2023-ml-w6p1/submit.csv')
print(submit.head())
```

```python
y_data = y_data['label']
test_data = test_data.drop('0', axis=1)
x_data = x_data.drop('0', axis=1)
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test=train_test_split(x_data,y_data, test_size=0.3, random_state=43, stratify=y_data)
```

```python
#LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(solver = 'svd', shrinkage=None, priors=None, n_components=None, store_covariance=None, tol=0.00001)
lda.fit_transform(x_data, y_data)
print(lda.score(X_test, Y_test))
```

```python
lda.fit(x_data, y_data)
model = lda.predict(test_data)
```

```python
submit['label'] = model
submit.to_csv('predict.csv', header=True, index = False)
```
