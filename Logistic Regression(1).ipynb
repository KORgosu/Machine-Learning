```python
train_X = pd.read_csv('/kaggle/input/2023-ml-w5p1/trainX.csv')

train_Y = pd.read_csv('/kaggle/input/2023-ml-w5p1/trainY.csv')

test_X = pd.read_csv('/kaggle/input/2023-ml-w5p1/testX.csv')

submit = pd.read_csv('/kaggle/input/2023-ml-w5p1/submit.csv')


```

```python
train_Y = train_Y['Category']
print(train_Y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_X, train_Y, test_size=0.3, random_state=100, stratify=train_Y)
```

```python
# Logistic Regression with L2 or L1 Regularization
from sklearn.linear_model import LogisticRegression
logist = LogisticRegression(C=200.0, max_iter = 1000)

l2_100 = LogisticRegression(penalty = 'l2', C= 100.0, solver = 'saga', max_iter=1000)
l2_10 = LogisticRegression(penalty='l2', C=10.0, solver='saga', max_iter = 1000)  # L2 with C(=1/λ)=10
l2_1 = LogisticRegression(penalty='l2', C=1.0, solver='saga', max_iter = 1000)    # L2 with C(=1/λ)=1
l2_0_1 = LogisticRegression(penalty='l2', C=0.1, solver='saga', max_iter = 1000)  # L2 with C(=1/λ)=0.1

l1_10 = LogisticRegression(penalty='l1', C=10.0, solver='saga', max_iter = 1000)  # L1 with C(=1/λ)=10
l1_1 = LogisticRegression(penalty='l1', C=1.0, solver='saga', max_iter = 1000)    # L1 with C(=1/λ)=1
l1_0_1 = LogisticRegression(penalty='l1', C=0.1, solver='saga', max_iter = 1000)  # L1 with C(=1/λ)=0.1
```

```python
logist.fit(X_train, y_train) # 페널티 디폴트
print('학습데이터 정밀도 with L2(default) and λ=0.005:', logist.score(X_train, y_train))
print('시험데이터 정밀도 with L2(default) and λ=0.005:', logist.score(X_test, y_test))

l2_100.fit(X_train, y_train)
print('학습데이터 정밀도 with L2 and λ=0.01:', l2_100.score(X_train, y_train))
print('시험데이터 정밀도 with L2 and λ=0.01:', l2_100.score(X_test, y_test))


l2_10.fit(X_train, y_train)
print('학습데이터 정밀도 with L2 and λ=0.1:', l2_10.score(X_train, y_train))
print('시험데이터 정밀도 with L2 and λ=0.1:', l2_10.score(X_test, y_test))

l2_1.fit(X_train, y_train)
print('학습데이터 정밀도 with L2 and λ=1:', l2_1.score(X_train, y_train))
print('시험데이터 정밀도 with L2 and λ=1:', l2_1.score(X_test, y_test))

l2_0_1.fit(X_train, y_train)
print('학습데이터 정밀도 with L2 and λ=10:', l2_0_1.score(X_train, y_train))
print('시험데이터 정밀도 with L2 and λ=10:', l2_0_1.score(X_test, y_test))

l1_10.fit(X_train, y_train)
print('학습데이터 정밀도 with L1 and λ=0.1:', l1_10.score(X_train, y_train))
print('시험데이터 정밀도 with L1 and λ=0.1:', l1_10.score(X_test, y_test))

l1_1.fit(X_train, y_train)
print('학습데이터 정밀도 with L1 and λ=1:', l1_1.score(X_train, y_train))
print('시험데이터 정밀도 with L1 and λ=1:', l1_1.score(X_test, y_test))

l1_0_1.fit(X_train, y_train)
print('학습데이터 정밀도 with L1 and λ=10:', l1_0_1.score(X_train, y_train))
print('시험데이터 정밀도 with L1 and λ=10:', l1_0_1.score(X_test, y_test))
```

```python
print(l2_10.intercept_)
print(l2_1.intercept_)
print(l2_0_1.intercept_)

print(l2_10.coef_)
print(l2_1.coef_)
print(l2_0_1.coef_)
```

```python
print(l1_10.intercept_)
print(l1_1.intercept_)
print(l1_0_1.intercept_)

print(l1_10.coef_)
print(l1_1.coef_)
print(l1_0_1.coef_)
```
