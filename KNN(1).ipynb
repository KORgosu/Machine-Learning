```python
import pandas as pd

# train 데이터 불러오기
train_df = pd.read_csv('/kaggle/input/2023-ml-w4p1/train.csv')
print(train_df)

# test 데이터 불러오기
test_df = pd.read_csv('/kaggle/input/2023-ml-w4p1/test.csv')
print(test_df)

# sample_submit 데이터 불러오
submit = pd.read_csv('/kaggle/input/2023-ml-w4p1/sample_submit.csv')
print(submit)
```

```python
# test 데이터에 label(y)값 드랍 및 확인
x = train_df.drop(['label'], axis = 1)
print(x.shape)
```

```python
# Y에 대한 정의 = label
y_data = train_df['label']
print(y_data)
```

```python
# 카테고리 실수화
from sklearn.preprocessing import LabelEncoder #LabelEncoder method
import numpy as np

classle = LabelEncoder()
y = classle.fit_transform(train_df['label'])
print('species label : ', np.unique(y)) # 중복되는 y값을 하나로 정리하여 print
```

```python
yo = classle.inverse_transform(y) #원래의 label 문자열로 전환
print('lebel:', np.unique(yo))
```

```python
# 데이터분할
from sklearn.model_selection import train_test_split #scikit-learn의 model_selection library를 train_test_split으로 명명

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1, stratify=y) #x, y를 각각 3:7의비율
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# KNN의 적용
from sklearn.neighbors import KNeighborsClassifier #KNN 불러오기
knn = KNeighborsClassifier(n_neighbors = 5, p = 2) # 5개의 인접한 이웃, 거리측정기준 : 유클리드
knn.fit(X_train, y_train) # 모델 피팅과정
```

```python
y_train_pred = knn.predict(X_train) #학습한 데이터 넣기
y_test_pred = knn.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, y_train_pred))
print(accuracy_score(y_test, y_test_pred))
```

```python
from sklearn.preprocessing import StandardScaler #Scikit-Learn의 model_selection library를 train_test_split으로 명명

std=StandardScaler()
std.fit(X_train) # 표준화
X_train_std=std.fit_transform(X_train) # normalized training data
X_test_std=std.transform(X_test) # nomarlized test data

print(X_train.head())
X_train_std[1:5, ]
```

```python
# KNN 적용
knn = KNeighborsClassifier(n_neighbors = 5, p=2)
knn.fit(X_train_std, y_train)
```

```python
y_train_pred_std = knn.predict(X_train_std)
y_test_pred_std = knn.predict(X_test_std)

print(accuracy_score(y_train, y_train_pred_std))
print(accuracy_score(y_test, y_test_pred_std))
```

```python
knn = KNeighborsClassifier(n_neighbors = 5, p = 2)
knn.fit(x, y) # 모델 검증
```

```python
y_label_pred = knn.predict(test_df)
y_label_pred_names = classle.inverse_transform(y_label_pred)
submit['label']= y_label_pred_names
submit.to_csv('submit.csv', mode='w', header= True, index= False)
```
