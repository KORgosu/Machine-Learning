```python
train_data = pd.read_csv('/kaggle/input/2023-ml-w6p2/train.csv')
print(train_data)
test_data = pd.read_csv('/kaggle/input/2023-ml-w6p2/test.csv')
print(test_data.head())
submit = pd.read_csv('/kaggle/input/2023-ml-w6p2/submit.csv')
print(submit.head())
```

```python
y_data = train_data['Diabetes']
x_data = train_data.drop('Diabetes', axis=1)
print(y_data.head())
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test=train_test_split(x_data,y_data, test_size=0.3, random_state=1)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit_transform(X_train, Y_train)
Y_train_pred = lda.predict(X_train)
Y_test_pred = lda.predict(X_test)
print(lda.score(X_test, Y_test))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_train, Y_train_pred)) # train data에 대한 accuracy
print(accuracy_score(Y_test, Y_test_pred)) # test data에 대한 accuracy
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train, Y_train)
Y_train_pred = qda.predict(X_train)
Y_test_pred = qda.predict(X_test)


print(accuracy_score(Y_train, Y_train_pred)) # train data에 대한 accuracy
print(accuracy_score(Y_test, Y_test_pred)) # test data에 대한 accuracy
print(qda.score(X_test, Y_test))
```

```python
qda.fit(X_train, Y_train)
model = qda.predict(test_data)
```

```python
submit['Diabetes'] = model
submit.to_csv('predict.csv', header=True, index = False)
```
